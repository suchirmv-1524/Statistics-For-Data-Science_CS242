```python


def extract_review_data(url):
    """Extracts customer review data from an Amazon product page, including multiple pages."""
    all_data = []
    page_number = 1
    max_pages = 30  # Adjust the maximum number of pages as needed

    while page_number <= max_pages:
        try:
            # Use headers to avoid blocking by Amazon's anti-scraping measures
            headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
            response = requests.get(url + f'&pageNumber={page_number}', headers=headers)
            soup = BeautifulSoup(response.content, 'html.parser')

            # Find customer review elements
            review_containers = soup.find_all('div', {'data-hook': 'review'})

            if not review_containers:
                print(f"No reviews found on page {page_number}.")
                break  # Stop if no reviews are found

            for review in review_containers:
                try:
                    customer_id = review.find('span', class_='a-profile-name').text.strip()
                    
                    # Extract rating from the 'review-star-rating' or 'a-icon-alt' class
                    rating_element = review.find('i', {'data-hook': 'review-star-rating'})
                    if not rating_element:
                        rating_element = review.find('span', {'class': 'a-icon-alt'})
                    
                    if rating_element:
                        rating = float(rating_element.text.split()[0])
                    else:
                        rating = 'NA'
                    
                    review_text = review.find('span', {'data-hook': 'review-body'}).text.strip()

                    data = {'customer_id': customer_id, 'rating': rating, 'review': review_text}
                    all_data.append(data)
                except Exception as e:
                    print(f"Error processing review: {e}")
                    continue

            # Check if there is a next page
            next_page = soup.find('li', class_='a-last')
            if next_page is None or 'a-disabled' in next_page['class']:
                break

            page_number += 1
        except Exception as e:
            print(f"Error processing page {page_number}: {e}")
            break  # Exit the loop if an error occurs

    if all_data:
        df = pd.DataFrame(all_data)
        return df
    else:
        print("Error: No reviews found on any pages.")
        return None

# Example usage
url = "https://www.amazon.in/Apple-iPhone-128GB-Deep-Purple/product-reviews/B0BDJ6ZMCC/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
reviews_df = extract_review_data(url)

if reviews_df is not None:
    print(reviews_df)
    reviews_df.to_csv('good_reviews.csv', index=False)
else:
    print("Error: Failed to extract reviews.")
```
